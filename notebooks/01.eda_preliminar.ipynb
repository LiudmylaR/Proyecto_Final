### Información sobre el conjunto de datos

La base de datos de ventas de comestibles es un conjunto de datos de 6 archivos:

1. **sales**:  Contiene datos transaccionales de cada venta.
 - SalesID:  Identificador único de cada venta.
 - SalesPersonID:  Empleado responsable de la venta.
 - CustomerID:  Cliente que realiza la compra.
 - ProductID:  Producto en venta.
 - Quantity:  Número de unidades vendidas.
 - Discount:  Descuento aplicado a la venta.
 - TotalPrice:  Precio final de venta después de descuentos. 
 - SalesDate:  Fecha y hora de la venta.
 - TransactionNumber:  Identificador único de la transacción.

2. **customers**:  Contiene información sobre los clientes que realizan compras.
 - CustomerID:  Identificador único para cada cliente.
 - FirstName:  Nombre del cliente.
 - MiddleInitial:  Inicial media del cliente.
 - LastName:  Apellido del cliente.
 - cityID:  Ciudad del cliente.
 - Address:  Dirección residencial del cliente.

4. **products**:  Almacena detalles sobre los productos que se venden.
 - ProductID:  Identificador único para cada producto.
 - ProductName:  Nombre del producto.
 - Price:  Precio por unidad del producto.
 - CategoryID:  Identificador de categoría único
 - Class:  Clasificación del producto.
 - ModifyDate:  Fecha de última modificación.
 - Resistant:  Categoría de resistencia del producto.
 - IsAllergic:  Indica si el artículo es un alérgeno
 - VitalityDays:  Clasificación de tipo vital de producto.

5. **categories**:  Define las categorías de los productos.
 - CategoryID:  Identificador único para cada categoría de producto.
 - CategoryName:  Nombre de la categoría de producto.

6. **cities**:  Contiene datos geográficos a nivel de ciudad.
 - CityID:  Identificador único para cada ciudad.
 - CityName:  Nombre de la ciudad.
 - Zipcode:  Población de la ciudad.
 - CountryID:  Referencia al país correspondiente.

7. **countries**:  Almacena metadatos relacionados con el país.
 - CountryID:  Identificador único para cada país.
 - CountryName:  Nombre del país.
 - CountryCode:  Código de país de dos letras.
