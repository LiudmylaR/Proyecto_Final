### Cargamos los conjuntos de datos desde un archivo CSV

```python
df_sales = pd.read_csv('../data/raw/sales.csv', parse_dates=['SalesDate'])
df_customers = pd.read_csv('../data/raw/customers.csv')
df_products = pd.read_csv('../data/raw/products.csv', parse_dates=['ModifyDate'])
df_categories = pd.read_csv('../data/raw/categories.csv')
df_cities = pd.read_csv('../data/raw/cities.csv')
df_countries = pd.read_csv('../data/raw/countries.csv')
```

### Creamos el dataset unico

Todos los archivos están vinculados y se pueden combinar en un único conjunto de datos.

Relaciones de datos:
- 'sales' --> 'products' (ProductID), 'customers' (CustomerID)
- 'customers' --> 'cities' (CityID)
- 'cities' --> 'countries' (CountryID)
- 'products' --> 'categories' (CategoryID)

Antes de unir los conjuntos de datos verificamos la integridad de las claves primarias y foráneas antes de realizar las uniones:
- ProductID
- CustomerID
- CityID
- CountryID
- CategoryID

Del análisis preliminar sabemos que no hay valores nulos ni duplicados entre las claves.

```python
se.claves_merge(df_sales, df_products,'ProductID')
```

```python
se.claves_merge(df_sales, df_customers,'CustomerID')
```

```python
se.claves_merge(df_customers, df_cities,'CityID')
```

```python
se.claves_merge(df_cities, df_countries,'CountryID')
```

```python
se.claves_merge(df_products, df_categories, 'CategoryID')
```

```python
# Al verificar las claves no se encontraron inconsistencias estructurales significativas.
# Podemos realizar uniones de las tablas
df = df_sales.merge(df_products, how = 'left', on = 'ProductID')
df = df.merge(df_customers, how = 'left', on = 'CustomerID')
df = df.merge(df_cities, how = 'left', on = 'CityID')
df = df.merge(df_countries, how = 'left', on = 'CountryID')
df = df.merge(df_categories, how = 'left', on = 'CategoryID')
```

```python
# Comprobamos el Dataframe unido:
df.head()
```

### Eliminación las columnas

```python
# Eliminemos las columnas que no son útiles para el análisis

columnas_no_utiles = [
    'ProductID',
    'TransactionNumber',
    'CategoryID',
    'ModifyDate',
    'FirstName',
    'MiddleInitial',
    'LastName',
    'CityID',
    'Address',
    'Zipcode',
    'CountryID',
    'CountryCode'
]

df = df.drop(columns=columnas_no_utiles)
```

### Gestion valores nulos

```python
se.porcent_nulos(df)
```

```python
# La variable SalesDate presentaba un 1% de valores nulos.
# Al ser clave para el análisis temporal y tener un impacto mínimo, 
# se eliminaron dichas observaciones para mantener la consistencia del análisis.

df = df.dropna(subset=['SalesDate'])
```

```python
df['SalesDate'].isnull().sum()
```

### Creación de variables nuevas

```python
# Antes de recalcular la columna 'TotalPrice' veamos el rango de columna 'Discount':
df['Discount'].describe()
```

```python
# Recalculamos la columna 'TotalPrice'
df['TotalPrice'] = df['Quantity'] * df['Price'] * (1 - df['Discount'])
df['TotalPrice'] = df['TotalPrice'].round(2)
df['TotalPrice'].describe()
```

```python
# Creamos la columna 'RevenueBruto' que permite analizar el impacto de los descuentos
df['RevenueBruto'] = df['Quantity'] * df['Price']
df['RevenueBruto'] = df['RevenueBruto'].round(2)
df['RevenueBruto'].describe()
```

```python
# Creamos la columna 'DiscountAmount'
df['DiscountAmount'] = df['RevenueBruto'] - df['TotalPrice']
df['DiscountAmount'] = df['DiscountAmount'].round(2)
df['DiscountAmount'].describe()
```

```python
# Creamos las columnas 'MonthNum', 'Month', 'DayOfWeek' para analisis de ventas temporal

df['MonthNum'] = pd.to_datetime(df['SalesDate']).dt.month
df['Month'] = pd.to_datetime(df['SalesDate']).dt.month_name()
df['DayOfWeek'] = pd.to_datetime(df['SalesDate']).dt.day_name()
```
